```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
datasets=pd.read_csv('insurance_pre (2).csv')
```

```python
datasets
```

```python
datasets=pd.get_dummies(datasets,drop_first=True)
```

```python
datasets
```

```python
indep=datasets[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dep=datasets[['charges']]
```

```python
indep
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=  StandardScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=50,random_state=0)
regressor.fit(X_train,Y_train)
```

```python
Y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import explained_variance_score
r_score=explained_variance_score(Y_test,Y_pred)
r_score
```
